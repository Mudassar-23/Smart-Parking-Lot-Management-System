1. Use dictionary to store

  key -> slot no.(s1-s5)
  value -> status("empty" or "occupied")

2. use tuple to store parking lot information
 parking-name, no.of solts  

3. use loop to display a menue

1. park vehicle                functions all
2. remove vehicle   
3.show parking status 
4. // summary     ------------> parking name,total slots, vehicles parked   
5. exit

4.  Maintain a list to store vehicle no.s that entered

////////////////////////////////////////////////////////

```python
# 2️. Tuple (Parking Info)
parking_info = ("Fast Parking", 5)

# 1️. Dictionary (Slots)
parking_slots = {'s1': "empty",'s2': "empty",'s3': "empty",'s4': "empty",'s5': "empty"}

# 4️. List (Vehicle Numbers)
vehicleslist = []


# 1. Park Vehicle
def park_vehicle():
    slot = input("Enter slot number (s1 to s5): ")
    
    if slot in parking_slots:
        if parking_slots[slot] == "empty":
            vehicle = input("Enter vehicle number: ")
            parking_slots[slot] = "occupied"
            
            vehicleslist.append(vehicle)
            print("Vehicle parked successfully!")
            summary() # Auto summary update
        else:
            print("Slot already occupied!")
    else:
        print("Invalid slot!")


# 2. Remove Vehicle
def remove_vehicle():
    slot = input("Enter slot number to remove vehicle: ")
    
    if slot in parking_slots:
        if parking_slots[slot] == "occupied":
            vehicle = input("Enter vehicle number: ")
            
            if vehicle in vehicleslist:
                vehicleslist.remove(vehicle)
                
                parking_slots[slot] = "empty"
                
                print("Vehicle removed successfully!")
                summary() # Auto summary update
            else:
                print("Vehicle not found!")
        else:
            print("Slot already empty!")
    else:
        print("Invalid slot!")


# 3. Show Parking Status
def show_status():
    print("\nParking Status:")
    for slot in parking_slots:
        print(slot, ": ", parking_slots[slot])


# 4. Summary
def summary():
    print("\n====== Parking Summary ======")
    print("Parking Name:", parking_info[0])
    print("Total Slots:", parking_info[1])
    print("Vehicles Parked:", len(vehicleslist))
    
    print("\nSlot Details:")
    for slot in parking_slots:
        print(slot, ":", parking_slots[slot])
    
    print("Vehicle Numbers:", vehicleslist)


# 3️. Menu Loop
while True:
    print("\n---- Parking Menu ----")
    print("1. Park Vehicle")
    print("2. Remove Vehicle")
    print("3. Show Parking Status")
    print("4. Summary")
    print("5. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        park_vehicle()
    elif choice == "2":
        remove_vehicle()
    elif choice == "3":
        show_status()
    elif choice == "4":
        summary()
    elif choice == "5":
        print("Exiting program...")
        break
    else:
        print("Invalid choice!")
```
